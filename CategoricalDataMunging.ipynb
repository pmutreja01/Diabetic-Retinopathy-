```python
import pandas as pd

data = pd.read_csv('D:/SEM-2/Project/DataCleaning/family_new.csv')
demo = pd.read_csv('D:/SEM-2/Project/DataCleaning/demographics-2.csv')
last_enc = pd.read_csv('D:/SEM-2/Project/DataCleaning/LatestEncounter4.csv')
last_enc

person_nbr = last_enc['Person_Nbr']
data['Rel'] = 1
demo['Demo'] = 1
```

```python
# JUST TO SHOW HOW WILL THIS PIVOT_TABLE WILL WORK ON THE FAMILY_HIST_FOR_ENC

(data.pivot_table(index='Person_Nbr', columns='Relation', values='Rel').fillna(0)).columns
```

```python
# JUST TO SHOW HOW WILL THIS PIVOT_TABLE WILL WORK ON THE FAMILY_HIST_FOR_ENC

data.pivot_table(index='Person_Nbr', columns='Relation', values='Rel').fillna(0)
```

```python
merged = pd.merge(left=data,right=last_enc, left_on='Person_Nbr', right_on='Person_Nbr',how='inner')
merged
```

```python
# CONVERTING A CATEGORICAL VARIABLE DEPENDING ON THE RELATION'S COLUMN UNIQUE VALUE
rel = pd.DataFrame
rel = merged.pivot_table(index='Person_Nbr', columns='Relation', values='Rel').fillna(0)
rel

#ab123 is the final dataframe that can be used to pass on to model but based on the encounter_nbr
```

```python
# CONVERTING A CATEGORICAL VARIABLE DEPENDING ON THE S'S COLUMN UNIQUE VALUE
smoking = pd.DataFrame
smoking = merged.pivot_table(index='Person_Nbr', columns='Smoking_Status', values='Rel').fillna(0)
smoking
```

```python
# MERGING THE DATAFRAME DEMOGRPAHICS AND LATEST_ENCOUNTER BASED ON PERSON_NBR
merged_inner = pd.merge(left=demo,right=last_enc, left_on='Person_Nbr', right_on='Person_Nbr',how='inner')
merged_inner
```

```python
# CREATING AGE FROM DOB AND THEN MAKING A DATA FRAME OF JUST AGE AND PERSON_NBR
from datetime import date
demographics=merged_inner
todayDate=date.today()
age=[]
#type(demographics['DOB'][1])
for i in range(len(demographics['DOB'])):
    
    d=pd.to_datetime(demographics['DOB'][i]).date()
    age.append(todayDate.year-d.year)
demographics['Age']=age
Age = merged_inner[['Person_Nbr','Age']]
Age
```

```python
# CONVERTING A CATEGORICAL VARIABLE DEPENDING ON THE RACE COLUMN UNIQUE VALUE
demo_race = pd.DataFrame
demo_race = merged_inner.pivot_table(index='Person_Nbr', columns='Race', values='Demo',aggfunc='sum').fillna(0)
demo_race
```

```python
# CONVERTING A CATEGORICAL VARIABLE DEPENDING ON THE ETHNICITY COLUMN UNIQUE VALUE
demo_eth = pd.DataFrame
demo_eth = merged_inner.pivot_table(index='Person_Nbr', columns='Ethnicity', values='Demo',aggfunc='sum').fillna(0)
demo_eth
```

```python
# A data frame merging both the race and ethncity based on person_nbr
demo_eth.reset_index( inplace=True)
demo_race.reset_index( inplace=True)

race_eth = pd.merge(demo_race,demo_eth, 'outer', on=['Person_Nbr'])
race_eth
```

```python
# MERGING THE RACE AND ETHNICITY WITH THE REL DATA FRAME BASED ON THE PERSON_NBR
# ALL THE COLUMNS LIKE RACE, ETHNICITY AND RELATION IS IN CATEGORICAL FORM
race_eth.reset_index( inplace=True)
rel.reset_index( inplace=True)
cat = pd.merge(race_eth,rel, 'outer', on=['Person_Nbr'])
cat
```

```python
# DELETING THE INDEX COLUMN
# x-race
# y-ethnicity
cat = cat.drop('index', 1)
cat
```

```python
# MERGING SMOKING STATUS TO THE OTHER CATEGORICALS VARIABLES
smoking.reset_index( inplace=True)
categorical = pd.merge(cat,smoking, 'outer', on=['Person_Nbr'])
categorical
```

```python
# MERGING THE AGE TO THE DATAFRAME THAT HAS ALL THE CATEGORICAL VARIBALES IN IT
Final = pd.merge(categorical,Age, 'outer', on=['Person_Nbr'])
Final
```

```python
Final=Final.rename(columns = {'Declined to specify_x':'Declined to specify_race','Declined to specify_y':'Declined to specify_ethnicity','Other':'Other_relations','Unknown':'Unknown_smoking_status','African American_x':'African American_race','African American_y':'African American_ethnicity'})
Final
```
